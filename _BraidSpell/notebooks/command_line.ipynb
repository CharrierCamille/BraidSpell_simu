# A. First introductive example using the command-line interface : reading the known word 'choisir' in French

if you don't need to store the data but just want a fast way to have a summary of the results, use the command-line interface.
If not provided, all parameters have the default value.

## 1. Usage

```sh
usage: braid [-h] [-n N] [-s S] [-m M] [-l L] [--lexicon LEXICON] [-f F] [-t T] [--fMin FMIN]
             [--maxItem MAXITEM] [--maxItemLen MAXITEMLEN] [--fMinPhono FMINPHONO]
             [--fMinOrtho FMINORTHO] [-v V] [-p P] [-g] [--Qa QA] [--sdA SDA] [--sdM SDM]
             [--thr_expo THR_EXPO] [--stop STOP] [--thr_fix THR_FIX] [--alpha ALPHA] [--debug]
             [--TD TD] [--TDPhono TDPHONO] [--phono PHONO] [--context CONTEXT] [--level LEVEL]
             [--time] [--fix_times FIX_TIMES] [--fix_pos FIX_POS]
             [--remove_stim_ortho] [--remove_stim_phono] [--remove_neighbors] [--remove_lemma]
             [--build_prototype] [--shift] [--explicit] [--semi-explicit] [--mixture_knowledge]
             [--n_choose_word N_CHOOSE_WORD]
```

If you don't want to use the installed version but rather use the source code (in case you modify it), the command to run in the project root directory becomes :
```sh
usage: python braid [-h] [-n N] [-s S] [-m M] [-l L] [--lexicon LEXICON] [-f F] [-t T] [--fMin FMIN]
             [--maxItem MAXITEM] [--maxItemLen MAXITEMLEN] [--fMinPhono FMINPHONO]
             [--fMinOrtho FMINORTHO] [-v V] [-p P] [-g] [--Qa QA] [--sdA SDA] [--sdM SDM]
             [--thr_expo THR_EXPO] [--stop STOP] [--thr_fix THR_FIX] [--alpha ALPHA] [--debug]
             [--TD TD] [--TDPhono TDPHONO] [--phono PHONO] [--context CONTEXT] [--level LEVEL]
             [--time] [--fix_times FIX_TIMES] [--fix_pos FIX_POS]
             [--remove_stim_ortho] [--remove_stim_phono] [--remove_neighbors] [--remove_lemma]
             [--build_prototype] [--shift] [--explicit] [--semi-explicit] [--mixture_knowledge]
             [--n_choose_word N_CHOOSE_WORD]
```
To get help about the command-line options, type :
```sh
braid -h
```

or

```sh
python braid -h
```

## 2. Default configuration

It's possible to run a simulation without giving any parameter, because like it was the case for running simulations with Python code,
there are default parameters to configure the simulation.

To run the simplest example where you choose your stimulus, for example the word "danser", type :

```sh
braid -s choisir
```

## 3. Data visualization

If you want to display plots corresponding to the simulation, type:
```sh
braid -s choisir -g
```

## 4. Some examples to configure the simulation

For example, if you want to simulate reading where the word is phonologically familiar but orthographically novel,
with a maximum of 1000 iterations, for 5 exposures, type:
```sh
braid -s choisir --remove_stim_ortho -m 1000 -n 5
```

If you want to simulate reading in English, type for example:
```sh
braid -l en -s help
```

If you want to get a lot of detailed information during simulation (at each iteration), type:
```sh
braid -s choisir --level debug
```

If you want a breakpoint at the end of the simulation to do some stuff, type
```sh
braid -s choisir --debug
```

in this case, everything can be accessed via the simu object 'sim', like sim.model, sim.res etc. (see other notebooks for more explanation)
